## Conjuntos

Un conjunto (`set`) es una colección **sin duplicados** y **sin orden**. Se crea con llaves `{}` o con `set()`.

Operaciones útiles:
- `|` unión
- `&` intersección
- `-` diferencia
- `in` para verificar pertenencia

## Diccionarios

Un diccionario (`dict`) guarda pares **clave: valor**. Se crea con `{clave: valor}`.

Operaciones útiles:
- `d[clave]` para acceder a un valor
- `d[clave] = valor` para agregar o modificar
- `d.keys()`, `d.values()`, `d.items()` para iterar
- `clave in d` para verificar si existe una clave
