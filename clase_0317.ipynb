```python
import turtle
caro = turtle.Turtle()
```

```python
caro.speed(5) # 1:slowest, 3:slow, 5:normal, 10:fast, 0:fastest
for i in range(4):
  caro.forward(100)  #avanzo 100 unidades
  caro.left(90)   # doblar 90 grados, por ejemplo a la izquierda

caro.penup()
caro.left(180)
caro.forward(20)

caro.pendown()
for i in range(4):
	caro.forward(100)
	caro.right(90)
```

```python
caro.penup()
caro.forward(120)
caro.pendown()
caro.penup()
caro.forward(100)
```

```python
caro.reset()
for k in range(3):
  #####################
  # dibuja un cuadrado y se corre
  #####################
  for j in range(3):
    for i in range(4):
      caro.forward(100)  #avanzo 100 unidades
      caro.left(90)   # doblar 90 grados, por ejemplo a la izquierda
    caro.penup()
    caro.forward(120)
    caro.pendown()
  #####################
  caro.penup()
  caro.backward(100+100+100+20+20+20)
  caro.right(90)
  caro.forward(120)
  caro.left(90)
  caro.pendown()

```
