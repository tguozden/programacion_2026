```python
import turtle
caro = turtle.Turtle()
```

```python
caro.speed(5) # 1:slowest, 3:slow, 5:normal, 10:fast, 0:fastest
for i in range(4):
  caro.forward(100)  #avanzo 100 unidades
  caro.left(90)   # doblar 90 grados, por ejemplo a la izquierda

caro.penup()
caro.left(180)
caro.forward(20)

caro.pendown()
for i in range(4):
	caro.forward(100)
	caro.right(90)
```
