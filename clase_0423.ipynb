```python
texto = "de qué color es el caballo blanco de san martín"
# supongamos que quiero saber cuántas letras d hay en el texto
cantidad_letras = dict()
#inicializo diccionario
for letra in texto:
    if letra != ' ':
        cantidad_letras[letra] = 0
    
# print(cantidad_letras)

for letra in texto:
    if letra is not ' ':
        cantidad_letras[letra] += 1
print(cantidad_letras)
```

```python
for clave in cantidad_letras:
    print(f"hay {cantidad_letras[clave]} caracteres {clave}")
```

```python
# histograma
for i in cantidad_letras:
    print(i, end=': ')
    for j in range(cantidad_letras[i]):
        print('*', end='')
    print()
```

```python
# cuál es el caracter que aparece más veces?
maximo = 0
# letra_max = texto[0] # lo podemos evitar
for i in cantidad_letras:
    if cantidad_letras[i] > maximo:
        maximo = cantidad_letras[i]
        letra_max = i 
print(letra_max, maximo)
```

```python
# a pedido de esteban resolvemos la meseta
#Meseta más larga:
# Escribí un programa que encuentre el índice 
# de inicio y fin de la meseta más larga de una lista. 
# Una meseta es una subsecuencia de elementos
# iguales consecutivos. Ejemplo: 
# lista = [1,3,3,1,1,7,7,7,7,4,5,5,4,4,4] 
# inicio lista[5] fin lista[8]

lista = [1,3,3,1,1,7,7,7,7,4,5,5,4,4,4] 
#       [0,1,0,1,0,3,2,1,0,0,1,0,2,1]   #lista de repeticiones
n = len(lista)
lista_rep=[]
for i in range( n - 1 ):   # recorro todos menos el último
    j = 0  # j es el índice de la subsecuencia
    while lista[i] == lista[i+j+1]:
        j += 1
        if i+j+1==n:
            break
    lista_rep.append(j)
```

```python
lista_rep
```

```python
imaximo = 0
for i in range(len(lista_rep)):
    if lista_rep[i]>lista_rep[imaximo]:
        imaximo = i
print(f"el máximo de lista_rep está en lista_rep[{imaximo}]")
        
    
```
