```python
# buscando máximos y mínimos
lista = [1, 2, 3, 56, 4]
maximo = lista[0]
for i in lista:
    if i> maximo:
        maximo = i
print(maximo)
```

```python
maximo = lista[0]
imaximo = 0
for i in range(len(lista)):
    if lista[i] > maximo:
        maximo = lista[i]
        imaximo = i
print(f'el máximo está en lista[{imaximo}]={maximo}')
```

```python

def dicc_de_listas(lista1, lista2):
    midicc = dict()
    for i in range(len(lista1)):
        midicc[ lista1[i] ] = lista2[i]  ##asigno un nuevo elemento al diccionario
    return midicc
lista1 = ['a', 'b', 'c', 'd']
lista2 = [2, 3, 4, 5, 6]
dicc_de_listas(lista1, lista2)
```
