## Anotaciones de tipos

En Python podemos indicar qué tipo de datos recibe y devuelve una función. Esto se llama **type hints** (pistas de tipo de datos). No cambia cómo funciona el programa, pero hace el código más claro.

```python
def saludar(nombre: str) -> str:
    return 'Hola, ' + nombre
```

- `nombre: str` → el parámetro `nombre` es un `str`
- `-> str` → la función devuelve un `str`

Tipo de datos que conosemos: `int`, `float`, `str`, `bool`, `list`, `dict`, `set`.

Si la función no devuelve nada (solo imprime), se anota `-> None`.

## Docstrings y `help()`

Un **docstring** es una cadena de texto que documenta una función. Se escribe justo después del `def`, entre triple comillas:

```python
def area_circulo(radio: float) -> float:
    """
    Calcula el área de un círculo dado su radio.

    radio: el radio del círculo (número positivo)
    Devuelve: el área como float
    """
    return 3.14159 * radio ** 2
```

Podés ver el docstring de cualquier función con `help()`:

```python
help(area_circulo)
help(len)
help(print)
```

## Diccionarios: más herramientas

Más métodos en diccionarios

```python
d = {'a': 1, 'b': 2}

d.get('a')          # 1  (igual que d['a'])
d.get('z')          # None  (no tira error si la clave no existe)
d.get('z', 0)       # 0   (si no existe devuelve 0)

d.update({'b': 99, 'c': 3})  # agrega más pares clave-valor (modifica "inplace")
```

Los diccionarios también pueden contener otros diccionarios como valores (**diccionarios anidados**):

```python
alumno = {
    'nombre': 'Ana',
    'nota': 8,
    'ciudad': 'Bariloche'
}
print(alumno['nombre'])   # Ana
```

Recordemos que sólo las claves tienen restricción sobre los tipos de datos. ¿Cuáles eran? ¿Por qué?
