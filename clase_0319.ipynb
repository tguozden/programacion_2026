# tipo de datos
# colecciones o compound data types

cumplen 3 cosas:

- puedo preguntar si un elemento está en la colección
- puedo preguntar cuántos elementos tiene
- puedo recorrer (iterar) todos los elementos

```python
lista = [1, 2, 3]
0 in lista  # in es pertenecer
```

```python
1 in lista
```

```python
len(lista)
```

```python
for i in lista:
    print(i)
```

```python
# tipos de colecciones
- listas
- strings
- tuplas
- conjuntos
- diccionarios
```

```python
l = [1, 2, 'abc', [1, 2, 3]] # listas
print(l)
```

```python
cadena = 'hola mundo!'  # string
print(cadena)
```

```python
mitupla = (1, 2, 'abc')   # esto es una tupla
print(type(mitupla).__name__)
print(type(mitupla))
1 + 2
```

```python
(1, 2, [1, 2,3])
```

```python
# tupla, lista y string
# son indexables
palabra = 'murciélago'
print(len(palabra))
print(palabra[0])
```

```python
# por ser indexable uno puede preguntar
# no solo si elemento está en el objeto
# sino además adónde está
palabra.find('i')
```

```python
# pregunta de lauti
z = 123
print(f"el caracter 3 está en la posición {str(z).find('3')}")
print("el caracter 3 está en la posición ", str(z).find('3'))
```
