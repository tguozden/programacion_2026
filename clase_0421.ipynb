```python
#estructuras de control


#for

for i in range(4):
    print(i, end = '-')
print()  # cambio de línea
for i in range(1,5):
    print(i, end = '-')
print()  # cambio de línea
for i in range(1, 10, 2):
    print(i, end = '-')
print()  # cambio de línea

cadena = 'hola mundo!'
for i in cadena:
    print(i, end = '-')
print()  # cambio de línea
```

```python
condicion = True
while condicion:
    print('hola! estoy en el scope del while')
    condicion = False
```

```python
temp = 15
if temp < 6:
    print('hace mucho frío')
elif temp >=6 and temp < 15:
    print('hace frío pero no tanto')
else:
    print('no hace frío')
```

```python
# break continue pass

for i in range(10):
    print(i, end='-')
    if i==4:
        break
```

```python
# continue:
for i in range(10):  
    if i==4:
        continue
    print(i, end='-')
```

```python
# pass
for i in range(4):
    pass
print('qué miran')
```
