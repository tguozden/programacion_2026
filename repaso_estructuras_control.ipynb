## `if` / `elif` / `else`

Ejecutan un bloque de código **según una condición**.

```python
if condicion_1:
    ...   # se ejecuta si condicion_1 es True
elif condicion_2:
    ...   # se ejecuta si condicion_1 es False y condicion_2 es True
else:
    ...   # se ejecuta si ninguna condición anterior fue True
```

- Puede haber **cero o más** `elif`.
- El `else` es opcional.
- Solo se ejecuta **el primer bloque** cuya condición sea `True`.

````python
# Ejemplo: clasificar una temperatura
temp = 22

if temp < 0:
    print('bajo cero')
elif temp < 10:
    print('frío')
elif temp < 25:
    print('agradable')
else:
    print('calor')

# Probá cambiando temp a -5, 5, 30
````

## `for`

Recorre **uno a uno** los elementos de una colección (lista, string, rango, diccionario...).

```python
for elemento in coleccion:
    ...   # se ejecuta una vez por cada elemento
```

Con `range()`:
```python
range(5)       # 0, 1, 2, 3, 4
range(2, 8)    # 2, 3, 4, 5, 6, 7
range(0, 10, 2)  # 0, 2, 4, 6, 8  (de a 2)
```

````python
# Ejemplo: sumar los números del 1 al 10
total = 0
for i in range(1, 11):
    total += i
print('Suma:', total)  # 55

# Ejemplo: recorrer una lista
frutas = ['manzana', 'pera', 'durazno']
for fruta in frutas:
    print(fruta.upper())
````

## `while`

Repite un bloque **mientras** una condición sea verdadera.

```python
while condicion:
    ...   # se repite mientras condicion sea True
```

⚠️ Si la condición nunca se vuelve `False`, el bucle no termina nunca (**bucle infinito**). Siempre asegurate de que algo dentro del bucle cambie el estado.

````python
# Ejemplo: dividir por 2 hasta llegar a menos de 1
x = 100
pasos = 0
while x >= 1:
    x = x / 2
    pasos += 1
print(f'Después de {pasos} divisiones: {x:.4f}')
````

## `break` y `continue`

Se usan **dentro de bucles** para alterar su comportamiento:

| Palabra | Efecto |
|---------|--------|
| `break` | **Sale** del bucle inmediatamente |
| `continue` | **Saltea** el resto de la iteración actual y pasa a la siguiente |

```python
for i in range(10):
    if i == 5:
        break       # sale cuando i llega a 5
    print(i)        # imprime 0, 1, 2, 3, 4

for i in range(10):
    if i % 2 == 0:
        continue    # saltea los pares
    print(i)        # imprime 1, 3, 5, 7, 9
```

````python
# Ejecutá y verificá que entendés la salida

print('--- break ---')
for i in range(10):
    if i == 5:
        break
    print(i, end=' ')
print()

print('--- continue ---')
for i in range(10):
    if i % 2 == 0:
        continue
    print(i, end=' ')
print()
````

## `pass`

`pass` es una instrucción que **no hace nada**. Se usa como marcador de posición cuando Python requiere un bloque de código pero todavía no queremos escribirlo.

```python
def funcion_pendiente():
    pass   # implementar después

for i in range(10):
    if i % 2 == 0:
        pass   # por ahora ignoramos los pares
    else:
        print(i)
```
