```python
# función misuma()
# toma una lista y suma todos los elementos
def misuma(listita):
    acumula = 0    # acumula es una variable local
    for i in listita:
        acumula += i
    print(acumula)    
```

```python
misuma([1, 2, 3])
```

```python
a = misuma([1, 2])
```

```python
b = sum([1, 2, 3])
print(b)
```

```python
def misuma2(listita):
    acumula = 0    # acumula es una variable local
    for i in listita:
        acumula += i
    return acumula
misuma2([2, 2, 2, 2]) + 2
```

```python
import turtle
x = turtle.Turtle()
```

```python
def dibuja_triangulo():
    for i in range(3):
        x.forward(20)
        x.left(120)
dibuja_triangulo()
```

```python
def dibuja_poligono(x, nlados, longlado):
    for i in range(nlados):
        x.forward(longlado)
        x.right(360/nlados)
```

```python
x.clear()
for i in range(20):
    dibuja_poligono(x, 4, 100)
    x.left(360/20)
```
