### Métodos `.sort()`, `sorted()` y `.reverse()`

- `.sort()` → ordena la lista **en el lugar** (modifica la original)
- `sorted()` → devuelve una lista nueva ordenada, **sin modificar** la original
- `.reverse()` → invierte la lista en el lugar

Estos métodos pueden recibir argumentos por nombre. Por ejemplo: `sorted(numeros, reverse=True)` ordena de mayor a menor. Cuando le pasás `reverse=True` estás usando lo que más adelante, cuando veamos funciones, se llama **keyword argument** (argumento por nombre).

1. Ordená `numeros` con `.sort()` e imprimí antes y después

3. Usá `sorted(numeros, reverse=True)` para obtener la lista ordenada de mayor a menor
4. Invertí la lista `letras` con `.reverse()`

```python
numeros = [5, 2, 8, 1, 9, 3]
letras = ['z', 'a', 'm', 'b', 'f']

# 1. .sort() modifica la lista original
print('Antes:', numeros)
numeros.sort()
print('Después de .sort():', numeros)

print('---')

# 2. sorted() devuelve una nueva lista
numeros2 = [5, 2, 8, 1, 9, 3]
nueva = sorted(numeros2)
print('sorted() devuelve:', nueva)
print('numeros2 sigue igual:', numeros2)  # ¿cambió?

print('---')

# 3. sorted() con argumento por nombre: reverse=True
numeros3 = [5, 2, 8, 1, 9, 3]
de_mayor_a_menor = sorted(numeros3, reverse=True)
print('De mayor a menor:', de_mayor_a_menor)

print('---')

# 4. .reverse()

print('Letras invertidas:', letras)
```

## Resumen de slicing

| Sintaxis | Significado |
|---|---|
| `s[i]` | elemento en posición `i` |
| `s[-1]` | último elemento |
| `s[a:b]` | desde `a` hasta `b` (sin incluir `b`) |
| `s[:b]` | desde el principio hasta `b` |
| `s[a:]` | desde `a` hasta el final |
| `s[::2]` | cada 2 elementos |
| `s[::-1]` | secuencia al revés |

Funciona igual para **listas**, **tuplas** y **strings**.

---
## Resumen de métodos de listas

| Método | Qué hace | Modifica la lista |
|---|---|---|
| `.append(x)` | agrega `x` al final | sí |
| `.extend(l)` | agrega todos los elementos de `l` | sí |
| `.remove(x)` | elimina la primera aparición de `x` | sí |
| `.pop(i)` | elimina y devuelve el elemento en posición `i` | sí |
| `.sort()` | ordena la lista | sí |
| `sorted(l)` | devuelve una nueva lista ordenada | no |
| `.reverse()` | invierte la lista | sí |
| `.count(x)` | cuenta las apariciones de `x` | no |
| `.copy()` | devuelve una copia independiente | no |
| `.index(x)` | devuelve la posición de `x` | no |
